```
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```
x_list = range(100)
df = pd.DataFrame(index=['x^2', '2*x^2', 'x^2+10^4', '2*x^2+10^4'], 
                  columns=x_list)
```

```
ｄｆ
```

```
for x in x_list:
    temp_list = [0]*4
    
    temp_list[0] = x**2
    temp_list[1] = 2*x**2
    temp_list[2] = x**2+10**4
    temp_list[3] = 2*x**2+10**4

    df[x] = temp_list
```

```
df
```

```
df_plot = df.T.plot()
plt.show()
```

```
df_plot = df.T.plot()
df_plot.legend(fontsize=15)
df_plot.set_xlabel('x', fontsize=18)
df_plot.set_ylabel('f(x)', fontsize=18)
df_plot.tick_params(labelsize=15)
plt.show()
```

```
df_plot = df.T.plot(style=['r-', 'r--', 'b-', 'b--'])
df_plot.legend(fontsize=15)
df_plot.set_xlabel('x', fontsize=18)
df_plot.set_ylabel('f(x)', fontsize=18)
df_plot.tick_params(labelsize=15)
plt.show()
```
